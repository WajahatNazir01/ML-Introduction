```python
X=data[:,0]
Y=data[:,1].reshape(X.size,1)
X = np.vstack((np.ones((X.size, )), X)).T
```

```python
print(X.shape)
print(Y.shape)
```

```python
plt.scatter(X[:,1],Y)
plt.show()
```

```python
def linearRegressionModel(X,Y,learningrate,iterations):
    m=Y.size
    #all zeors in theta in start theta represents features here
    thetaW=np.zeros((2,1))
    cost_list=[]
    for i in range(iterations):
        y_pred=np.dot(X,thetaW)
        cost=(1/(2*m))*np.sum(np.square(y_pred-Y))
        der_thetaW=(1/m)*np.dot(X.T,y_pred-Y)
        thetaW=thetaW-(learningrate*der_thetaW)

        cost_list.append(cost)
    return thetaW,cost_list
        
```

```python
iterations=9
learningrate=0.00000005
thetaW,cost_list=linearRegressionModel(X,Y,learningrate,iterations)
```

```python
new_houses = np.array([[1, 1547], [1, 1896], [1, 1934], [1,
2800], [1, 3400], [1, 5000]])
for house in new_houses :
    print("Our model predicts the price of house with",
    house[1], "sq. ft. area as : $", round(np.dot(house, thetaW)[0],
    2))
```

```python
rng = np.arange(0,iterations)
plt.plot(cost_list,rng)
plt.show()
```
